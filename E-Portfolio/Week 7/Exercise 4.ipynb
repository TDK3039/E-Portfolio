```python
import random
import matplotlib.pyplot as plt
import numpy as np

def slotted_aloha_Simulation(num_Nodes, transit_Prob, num_Slots):
    successful_Transmissions = 0 
    for _ in range(num_Slots):
        Transmissions = sum(1 for _ in range(num_Nodes) if random.random() < transit_Prob)
        if Transmissions == 1:
            successful_Transmissions += 1
            
    return successful_Transmissions / num_Slots
    
nodes = 50
probabilities = np.linspace(0, 1, 50)
efficiency = [slotted_aloha_Simulation(nodes, p, 10000) for p in probabilities]

plt.plot(probabilities, efficiency, marker='o')
plt.xlabel("The Transmission Probablity")
plt.ylabel("Efficiency")
plt.title("Slotted ALOHA Efficiency")
plt.grid()
plt.show
            
            
```
